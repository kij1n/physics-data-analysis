```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_all, plot_each, add_resonance_plot_labels, ResonanceFitData, PendulumsData
```

```python
# lengths = np.array([0.97, 0.99, 0.995, 1.0])
# labels = ["97cm", "99cm", "99.5cm", "100cm"]
lengths = np.array([1.01, 1.0, 0.995, 0.99])
labels = ["101cm", "100cm", "99.5cm", "99cm"]
cols = ["mass A", "mass B", "mass C", "mass D"]
L_d = 1.0
```

```python
increased_mass = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/increased-mass.csv").dropna()
)
doubled_mass = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/doubled-mass.csv").dropna()
)
short_dress = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/short-skirt.csv").dropna()
)
long_dress = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/long-skirt.csv").dropna()
)
```

```python
# drop range to skip transient phase
t_start = 0
t_end = 300
increased_mass.drop_time_range(t_start, t_end)
doubled_mass.drop_time_range(t_start, t_end)
short_dress.drop_time_range(t_start, t_end)
long_dress.drop_time_range(t_start, t_end)
```

```python
increased_mass_fit = ResonanceFitData(L_d, increased_mass)
doubled_mass_fit = ResonanceFitData(L_d, doubled_mass)
short_dress_fit = ResonanceFitData(L_d, short_dress)
long_dress_fit = ResonanceFitData(L_d, long_dress)
```

## Increased mass
Driver's mass was increased to 600g and pendulums' to 100g each.

```python
plot_all(increased_mass, title="Increased Mass")
plt.show()
plot_each(increased_mass, title="Increased Mass")
plt.show()

increased_mass_fit.fit()
increased_mass_fit.plot_detected_peaks(title="Increased Mass detected peaks")
print("Increased Mass detected peaks:")
print(increased_mass_fit.envelope_data)
plt.show()
increased_mass_fit.plot_resonance_curve("red", "black")
add_resonance_plot_labels(title="Increased Mass resonance curve")
plt.show()
```

## Doubled mass
Driver's mass was doubled to 1 kg, pendulums' masses stayed 100 g each

```python
plot_all(doubled_mass, title="Doubled Mass")
plt.show()
plot_each(doubled_mass, title="Doubled Mass")
plt.show()

doubled_mass_fit.fit()
doubled_mass_fit.plot_detected_peaks(title="Doubled Mass detected peaks")
print("Doubled Mass detected peaks:")
print(doubled_mass_fit.envelope_data)
plt.show()
doubled_mass_fit.plot_resonance_curve("red", "black")
add_resonance_plot_labels(title="Doubled Mass resonance curve")
plt.show()
```

## Short Dress
The masses were reverted to the previous values (500 g for the driver, 50 g for the pendulums). To each of the pendulums was attached a paper wrap (hence, dress) to increase air resistance. The dress's length was 7 cm.

```python
plot_all(short_dress, title="Short Dress")
plt.show()
plot_each(short_dress, title="Short Dress")
plt.show()

short_dress_fit.fit()
short_dress_fit.plot_detected_peaks(title="Short Dress detected peaks")
print("Short Dress detected peaks:")
print(short_dress_fit.envelope_data)
plt.show()
short_dress_fit.plot_resonance_curve("red", "black")
add_resonance_plot_labels(title="Short Dress resonance curve")
plt.show()
```

## Long dress
The masses stayed as with short dress example. The dresses were extended to 10.5 cm.

```python
plot_all(long_dress, title="Long Dress")
plt.show()
plot_each(long_dress, title="Long Dress")
plt.show()

long_dress_fit.fit()
long_dress_fit.plot_detected_peaks(title="Long Dress detected peaks")
print("Long Dress detected peaks:")
print(long_dress_fit.envelope_data)
plt.show()
long_dress_fit.plot_resonance_curve("red", "black")
add_resonance_plot_labels(title="Long Dress resonance curve")
plt.show()
```
