```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_all, plot_each, add_resonance_plot_labels, ResonanceFitData, PendulumsData, save_plot
```

```python
lengths = np.array([0.97, 0.99, 0.995, 1.0])
labels = ["97cm", "99cm", "99.5cm", "100cm"]

# tests
# lengths = np.array([1.01, 1.0, 0.995, 0.99])
# labels = ["101cm", "100cm", "99.5cm", "99cm"]

cols = ["mass A", "mass B", "mass C", "mass D"]
L_d = 1.0
```

```python
increased_mass = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/increased-mass.csv").dropna()
)
doubled_mass = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/doubled-mass.csv").dropna()
)
short_dress = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/short-skirt.csv").dropna()
)
long_dress = PendulumsData(
    cols, labels, lengths, 
    pd.read_csv("../data/session-6/long-skirt.csv").dropna()
)
```

```python
# remove origin axis differences
increased_mass.preprocess_offsets()
doubled_mass.preprocess_offsets()
short_dress.preprocess_offsets()
long_dress.preprocess_offsets()
```

```python
# drop range to skip transient phase
t_range = (0, 300)
increased_mass.drop_time_range(*t_range)
doubled_mass.drop_time_range(*t_range)
short_dress.drop_time_range(*t_range)
long_dress.drop_time_range(*t_range)
```

```python
increased_mass_fit = ResonanceFitData(L_d, increased_mass)
doubled_mass_fit = ResonanceFitData(L_d, doubled_mass)
short_dress_fit = ResonanceFitData(L_d, short_dress)
long_dress_fit = ResonanceFitData(L_d, long_dress)
```

## Increased mass
Driver's mass was increased to 600g and pendulums' to 100g each.

```python
# base plots
plot_all(increased_mass, title="Increased Mass")
save_plot("session-6/amp-vs-time-increased-mass-all.svg")
plt.show()

plot_each(increased_mass, title="Increased Mass")
save_plot("session-6/amp-vs-time-increased-mass-each.svg")
plt.show()

# detected peaks
increased_mass_fit.fit()
increased_mass_fit.plot_detected_peaks(title="Increased Mass detected peaks")
save_plot("session-6/detected-peaks-increased-mass.svg")
plt.show()


print("Increased Mass detected peaks:")
print(increased_mass_fit.envelope_data)

# resonance curve
increased_mass_fit.plot_resonance_curve("red", "black")
add_resonance_plot_labels(title="Increased Mass resonance curve")
save_plot("session-6/resonance-curve-increased-mass.svg")
plt.show()
```
